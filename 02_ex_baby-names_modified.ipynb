## TRỰC QUAN DỮ LIỆU

### 09. Vẽ đồ thị top 10 tên xuất hiện nhiều nhất năm 2014

```python
z = baby_names[baby_names.Year == 2014]
z
```

```python
t = z.groupby('Name').size().sort_values(ascending=False).head(10)
t
```

```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
#plt.subplots(figsize=(12, 4))
plt.plot(t.index, t.values)
plt.show()
```
