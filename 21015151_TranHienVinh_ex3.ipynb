## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

````python
import numpy as np
import matplotlib.pyplot as plt
arr=np.random.randint(1000, size=100)
print(arr)
````

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

````python
print(f"10 phần tủ đầu tiên: {arr[0:10]}")
print(f"5 phần tử cuối cùng: {arr[-5:]}")
elements=[i for i in arr if i%5==0 and i>100]
print(f"Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100: {elements}")
````

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

````python
elements_below_100=[item for item in arr if item <100]
print(f"Mảng trên có {len(elements_below_100)} phần tử có giá trị nhỏ hơn 100")
print(f"Các phần tử có giá trị nhỏ hơn 100: {elements_below_100}")
````

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

````python
print(f"Sum array: {arr.sum()}")
print(f"Mean: {arr.mean()}")
print(f"Median: {np.median(arr)}")
print(f"Quartile[Q1, Q2, Q3]: {np.quantile(arr,[.25,.5,.75])}")
print(f"Percentile 20: {np.percentile(arr,20)}")
````

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

````python
from numpy import random
for i in range(5):
    random_selection = random.choice(arr, 10, replace=False)
    print(f"Lần {i+1}: {random_selection}")
````

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

````python
unique_values, counts=np.unique(arr, return_counts=True)
index_of_most_common = np.argmax(counts)
most_common_element = unique_values[index_of_most_common]
print(f"Phần tử xuất hiện nhiều nhất: {most_common_element}")
````

### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

````python
# Your code here
plt.subplots(figsize=(9,7))
x=np.linspace(0, np.pi*2)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1, label="sin(x)")
plt.plot(x,y2, label="cos(x)", ls="--")
plt.title("Đồ thị hàm sin(x) và cos(x)")
plt.grid(True)
plt.legend()
plt.show()
````

## SEABORN



Seaborn là một thư viện hỗ trợ trực quan hóa được viết dựa trên thư viện `matplotlib`. Trong bài tập này, bạn hãy tự tìm hiểu thư viện `seaborn` và trực quan hóa dữ liệu tiền tip tại một nhà hàng.

Bạn có thể dùng đoạn mã sau để đọc dữ liệu có sẵn trong `seaborn`

```python
import seaborn as sns

df = sns.load_dataset("tips")
df.head()
```
Hoặc bạn có thể đọc dữ liệu chứa trong file `tips.csv` chứa thông tin tiền tip của khách hàng tại một nhà hàng. Dữ liệu có các thông tin sau:

- `total_bill`: tổng hoá đơn (USD)
- `tip`: tiền tip/ tiền boa (USD)
- `sex`: giới tính người thanh toán hóa đơn
- `smoker`: có hút thuốc hay không
- `day`: ngày trong tuần
- `time`: thời gian dùng bữa(buổi sáng, buổi trưa, buổi tối)
- `size`: số lượng người tham gia

Bạn hãy lựa chọn loại đồ thị phù hợp để trả lời những câu hỏi sau:

### 1. Vẽ đồ thị thể hiện tổng tiền tip theo ngày

````python
import seaborn as sns
import matplotlib.pyplot as plt
# Đọc dữ liệu từ file tips
tips = sns.load_dataset("tips")

# Gom nhóm theo ngày
group_by_day=tips.groupby(["day"], observed=False)
# Lấy cột tip trong gom nhóm theo ngày và tính tổng
tip_sum_by_day=group_by_day['tip'].sum().reset_index()

# Vẽ đồ thị sử dụng seaborn
plt.figure(figsize=(8, 6))
barplot = sns.barplot(x="day", y="tip", data=tip_sum_by_day, palette="inferno", hue="day")
plt.title("Tổng tiền tip theo ngày")
plt.xlabel("Ngày")
plt.ylabel("Tổng tiền tip")
plt.show()
````

### 2. Vẽ đồ thị thể hiện tổng tiền tip theo thời gian dùng bữa (time)


````python
# Gom nhóm thời gian dùng bữa
group_by_time=tips.groupby(['time'],observed=False)
# Lây cột tip trong gom nhóm theo day và tính tổng
tip_sum_by_time=group_by_time['tip'].sum().reset_index()
# Vẽ sơ đồ bằng seaborn
plt.figure(figsize=(8,6))
barplot=sns.barplot(x='time', y='tip',data=tip_sum_by_time, palette="inferno", hue='time')
plt.xlabel("Time")
plt.ylabel("Tổng tiền tip")
plt.show()
````

### 3. Vẽ đồ thị histogram của tổng hoá đơn


````python
histplot=sns.histplot(data=tips, x="total_bill", bins=20)
plt.title("Tổng giá đơn")
plt.xlabel("Tổng hóa đơn")
plt.ylabel("Tần xuất")
plt.show()
````

### 4. Vẽ đồ thị boxplot của tổng hoá đơn


````python
data=tips['total_bill']
boxplot=sns.boxplot(data=data)
plt.title("Tổng hóa đơn")
plt.show()
````

### 5. Vẽ đồ thị boxplot so sánh tổng hóa đơn giữa các ngày trong tuần


````python
boxplot=sns.boxplot(data=tips, x='day', y='total_bill')
plt.title("So sánh tổng hóa đơn giữa các ngày trong tuần")
plt.xlabel("Ngày")
plt.ylabel("Tổng háo đơn")
plt.show()
````

### 6. Vẽ đồ thị để kiểm tra sự liên quan giữa tổng hoá đơn và tiền tip. (`scatter plot`)


````python
scatter_plot = sns.scatterplot(x="total_bill", y="tip", data=tips, color="blue")
plt.title("Scatter plot giữa Tổng hóa đơn và Tiền Tip")
plt.xlabel("Tổng hóa đơn")
plt.ylabel("Tiền Tip")
plt.show()
````

### 7. Tìm hiểu và sử dụng hàm `jointplot` trong seaborn để khảo sát quan hệ giữa hai thuộc tính `total_bill` và `tip`. Theo bạn hai thuộc tính này có tương quan với nhau hay không?


````python
joint_plot = sns.jointplot(x="total_bill", y="tip", data=tips, color="b")
plt.suptitle("Quan hệ giữa Tổng hóa đơn và Tiền Tip")
plt.show()
````

### 8. Tìm hiểu và sử dụng hàm `pairplot` trong seaborn để trực quan bộ dữ liệu. Theo bạn, trong bộ dữ liệu có những thuộc tính nào liên quan với nhau không?



````python
pairplot=sns.pairplot(tips)
plt.show()
````

### 9. Vẽ hai đồ thị histogram cạnh nhau để so sánh tổng hóa đơn giữa hai thời gian dùng bữa `Lunch` và `Dinner`



````python
fig,ax=plt.subplots(1,2, figsize=(9,6))
sns.histplot(data=tips[tips['time']=='Lunch']['total_bill'], bins=20, ax=ax[0])
sns.histplot(data=tips[tips['time']=='Dinner']['total_bill'], bins=20, ax=ax[1])
ax[0].set_title("Tổng hóa đơn bữa Lunch")
ax[0].set_xlabel("Tổng bill")
ax[0].set_ylabel("Số lượng")
ax[1].set_title("Tổng hóa đơn bữa Dinner")
ax[1].set_xlabel("Tổng bill")
ax[1].set_ylabel("Số lượng")
plt.show()
````

### 10. Vẽ hai đồ thị boxplot cạnh nhau để so sánh tổng hóa đơn giữa hai thời gian dùng bữa `Lunch` và `Dinner`

````python
plt.figure(figsize=(10, 6))
boxplot = sns.boxplot(x='time', y='total_bill', data=tips, palette=['skyblue', 'salmon'])
plt.title('Tổng hóa đơn theo time')
plt.xlabel('Time')
plt.ylabel('Tổng hóa đơn')
plt.show()
````
